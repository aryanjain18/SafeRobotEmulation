```python
import numpy as np
import os
import matplotlib.pyplot as plt
CWD = os.getcwd()
```

```python
def plot(title, data):
    mean = data.mean()
    std = data.std()
    plt.plot(data)
    plt.title(f"{title}: Mean-{mean} : Std.: {std}")
    plt.xlabel("packet No->>")
    plt.ylabel("delay-->")
    plt.show()
```

```python
def getdata(file):
    with open(CWD + "\\" + file, "r") as f:
        data = f.readlines()

    data = [x.strip().split(",") for x in data]
    head = data[0]
    data = np.array([[float(x) for x in row] for row in data[1:]])

    for i in range(1, len(head)):
        title = head[i]
        col = data[:, i]
        plot(title, col)
```
